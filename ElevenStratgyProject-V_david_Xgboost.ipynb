```python
import os
import csv
import requests
from bs4 import BeautifulSoup
import pandas as pd

requete = requests.get("https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20170325&end=20190325") #start date and end date in url
page = requete.content
soup = BeautifulSoup(page)

tbl = soup.find('table')
rows = tbl.find_all('tr')
bitcoin = pd.DataFrame(columns=['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Cap'])
i = 0
for row in rows[1:]: #first row is header so we ignore it
    value = row.find_all('td')
    liste = [val.get_text() for val in value]
    bitcoin.loc[i] = liste
    i = i + 1
```

```python
bitcoin['Open'] = bitcoin['Open'].astype(float)
```

```python
bitcoin['High'] = bitcoin['High'].astype(float)
```

```python
bitcoin['Low'] = bitcoin['Low'].astype(float)
```

```python
bitcoin['Close'] = bitcoin['Close'].astype(float)
```

```python
bitcoin['diffClose']=bitcoin['Close']
```

```python
bitcoin.dtypes
```

```python
bitcoin['diffClose']=bitcoin.diffClose.pct_change() * 100
```

```python
bitcoin["Volume"] = bitcoin["Volume"].str.replace(",","").astype(float)
```

```python
bitcoin['diffVolume']=bitcoin["Volume"]
```

```python
bitcoin['diffVolume']=bitcoin.diffVolume.pct_change() * 100
```

```python
df4=pd.read_csv("C:/Users/david/Desktop/Master in Data Science/Cours/projeteleven3.csv", sep=',')
```

```python
df4["up_down"] = 0
for i in range(len(df4["diffVolume"])):
    if df4.diffVolume[i] > 0 :
        df4.up_down[i] = 1
```

```python
df4['Date'] = df4['Date'].astype(str)
df4['new_date'] = df4['Date'].str.replace('\D', '').astype(int)
```

```python
df4 = df4.drop(columns=["Cap","Date","diffClose","diffVolume"])
```

```python
df4.to_csv(r'C:/Users/david/Desktop/monzizi.csv',index= False)
```

```python
X = df4.drop(columns=['up_down'])
y = df4.up_down
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#XGboost
xgb = XGBClassifier()
parameters = dict(model__max_depth=[1,3,5],
                  model__learning_rate=[.001,0.05, .1],
                  model__n_estimators=[50, 150])

                     
grid_search = GridSearchCV(xgb,param_grid = parameters,scoring='accuracy', verbose=1, cv=10)
```

```python
grid_search.fit(X_train,y_train)

# make predictions for test data
y_pred = grid_search.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```
